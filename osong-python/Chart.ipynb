```python
squares = [num**2 for num in range(1,100)]
squares
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(squares)

plt.show()
```

```python
plt.style.use("seaborn-v0_8-bright")

x_values = [x for x in range(1, 31)]
y_values = [y**2 for y in range(1, 31)] # squares

fig, ax = plt.subplots()
ax.scatter(x_values, y_values, c = y_values, cmap = plt.cm.tab20, s = 100)

plt.show()
```

### Color maps (ALL)

```python
import matplotlib as mlp
import matplotlib.pyplot as plt

def plot_cmaps(cmap):
    fig, ax = plt.subplots(figsize = (4, 0.4))
    col_map = plt.get_cmap(cmap)
    mlp.colorbar.ColorbarBase(ax, cmap = col_map, orientation = 'horizontal')
    
    plt.show()
    
for cmap_id in plt.colormaps():
    print(cmap_id)
    plot_cmaps(cmap_id)
```
