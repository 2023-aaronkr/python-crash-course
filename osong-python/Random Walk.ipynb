# Random Walk

**랜덤 워크**란 일련의 단순하면서도 랜덤한 결정에 따라 경로를 만드는 걸 말합니다.

```python
from random import choice

class RandomWalk:
    '''랜덤 워크를 만드는 클래스'''
    
    def __init__(self, num_pts = 5000):
        '''속성을 초기화합니다'''
        self.num_pts = num_pts
        
        # 이동은 (0,0)에서 시작합니다
        self.x_values = [0]
        self.y_values = [0]
        
    def fill_walk(self):
        '''랜덤 워크의 각 포인트를 계산합니다'''
        
        # 설정한 이동 수에 도달할 때까지 움직임을 반복합니다
        while len(self.x_values) < self.num_pts:
            
            # 방향과 거리를 정합니다
            x_dir = choice([1, -1]) # 1 = 오른쪽, -1 = 왼쪽
            x_dist = choice([0,1,2,3,4]) # x 방향의 거리 
            x_step = x_dir * x_dist
            
            y_dir = choice([1, -1]) # 1 = 위쪽, -1 = 아래쪽
            y_dist = choice([0,1,2,3,4]) # y 방향의 거리 
            y_step = y_dir * y_dist
            
            if x_step == 0 and y_step == 0:
                continue
                
            # 새 위치를 계산합니다
            x = self.x_values[-1] + x_step
            y = self.y_values[-1] + y_step
            
            self.x_values.append(x)
            self.y_values.append(y)
```

```python
import matplotlib.pyplot as plt

# 랜덤 워크를 생성합니다
rw = RandomWalk(5_0000) # 초기화 (50,000 점)
rw.fill_walk() # 모든 포인트 계산

# 랜덤 워크의 포인트를 그립니다
plt.style.use('classic')

fig, ax = plt.subplots()
pt_nums = range(rw.num_pts)

ax.scatter(rw.x_values, rw.y_values, 
           c = pt_nums, cmap = plt.cm.Greens, 
           edgecolors = 'none',
           s = 10)
ax.set_aspect('equal')

# 시작점과 끝점을 강조합니다
ax.scatter(0, 0, c = 'blue', edgecolors = 'none', s = 100) # 시작점
ax.scatter(rw.x_values[-1], rw.y_values[-1], 
           c = 'red', edgecolors = 'none', s = 100) # 끝점

# 축을 제거합니다
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

# 이미지 저장하기
plt.savefig('random_walk.png', bbox_inches = 'tight')

plt.show()
```
